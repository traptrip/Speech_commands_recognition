```python
from pathlib import Path
import librosa

hackaton_ds = Path("hackaton_ds")
train = hackaton_ds/"train"
test = hackaton_ds/"test"
noises = hackaton_ds/"noises"
```

```python
for directory in train.iterdir():
    print(directory, len(list(directory.glob("*.wav"))))
```

```python
for directory in noises.iterdir():
    print(directory, len(list(directory.glob("*.wav"))))
```

```python
len(list(test.glob("*.wav")))
```

```python
def aver(l: list):
    return sum(l) / len(l)
```

```python
duratinos = list()
for wav_file in list(test.glob("*.wav")):
    duratinos.append(librosa.get_duration(filename=wav_file))
print(min(duratinos))
print(aver(duratinos))
print(max(duratinos))
```
